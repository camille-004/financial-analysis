# Load Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import datetime

# Visualization
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (5, 4)

from pandas_datareader import data as pdr
yf.pdr_override()

start = datetime.datetime(2000, 2, 1)
end = datetime.datetime(2020, 2, 1)

tickers = [
    'AAPL', 'AMGN', 'CSCO', 'CVX', 'DIS', 
    'HD', 'HON', 'IBM', 'INTC', 'JNJ', 'JPM', 
    'KO', 'MCD', 'MRK', 'MSFT', 'NKE', 'PG', 
    'UNH', 'VZ', 'WMT'
]

stocks = pdr.DataReader(tickers, start, end, interval='1mo')['Adj Close'].dropna()
stocks.index = stocks.index.normalize() - pd.offsets.MonthBegin(1)
stocks
```

## Read market data

```python
market = pd.read_csv('^GSPC.csv')['Adj Close']
stocks['SNP'] = market.values
stocks
```

# 1. Calculate monthly returns
Calculated using percent change from previous monthly price, given by $Ret=[P(t)/P(t-1)]-1$.

```python
monthly_returns = stocks.pct_change() * 100
monthly_returns
```

## Get market risk premium and and risk free rate data
We have to get the most recent 25 years instead of the most recent 20 so we won't have missing values for the 60-month rolling betas.

```python
risk_data = pd.read_csv('F-F_Research_Data_Factors.CSV')
risk_data = risk_data.set_index('date')
risk_data.index = pd.to_datetime(risk_data.index, format='%Y%m', errors='coerce')
risk_data = risk_data[(
    risk_data.index >= datetime.datetime(2000, 9, 1)) & (
    risk_data.index <= datetime.datetime(2019, 9, 1))]
risk_data.drop(risk_data.tail(1).index, inplace=True)
risk_data
```

# 2. Get Rolling Betas

```python
def rolling_rise_over_run(ticker):
    return (monthly_returns[ticker].diff() / monthly_returns['SNP'].diff()).rolling(60).mean()

betas = pd.DataFrame()

for ticker in tickers:
    col = rolling_rise_over_run(ticker)
    betas[ticker] = col
    
# First 60 values would be NaN
betas = betas.iloc[60:]

# One value will still be NaN since the first row was, in monthly_returns
betas
```

# 3. Monthly Quintiles

```python
quintile_list = []

for i in range(len(betas)):
    quintiles = pd.qcut(betas.iloc[i], 5, labels=False, duplicates='drop').values
    quintile_list.append(quintiles)
    
quintile_df = pd.DataFrame(
    quintile_list, columns=tickers, index=betas.index).fillna(-1)
quintile_df = quintile_df.astype(int) + 1
quintile_df
```

# 4. Long-Short Trading Strategy
## Monthly portfolio returns

```python
# Create empty DataFrame with zeros
top = 4
bottom = 4
df_top = pd.DataFrame(0, index=betas.index, columns=betas.columns)
df_bottom = pd.DataFrame(0, index=betas.index, columns=betas.columns)

# Finding indicators of top 4 (top quintile) and bottom 4
# (bottom quintile) performing stocks on given month
for index, column in betas.iterrows():
    df_top.loc[index, betas.loc[index].nsmallest(top).index] = 1
    df_bottom.loc[index, betas.loc[index].nlargest(bottom).index] = 1
    
# Compute portfolio returns
portfolio_returns = (
    monthly_returns[df_top == 1].fillna(0).rolling(1).mean() - monthly_returns[
        df_bottom == 1].fillna(0).rolling(1).mean())
monthly_portfolio_returns = portfolio_returns.T.sum()
monthly_portfolio_returns[60:]
```

```python
plt.plot(monthly_portfolio_returns[60:])
```

## Compute Statistics

```python
# Mean
returns_mean = monthly_portfolio_returns.mean()

# Standard Deviation
returns_sd = monthly_portfolio_returns.sem()

print(f'Mean: {returns_mean}, Standard Deviation: {returns_sd}')
```

## Conduct One-Sample T-Test
Check if the actual mean from the generated signal is significantly greater than zero.

```python
from scipy import stats
import scipy as sc

t, p = sc.stats.ttest_1samp(monthly_portfolio_returns, 0)
print('T-statistic = {}, p-value = {}'.format(t, p/2))
```

We obtained `2.082` for the $t$-statistic and a $p$-value less than 0.05. Therefore, we can reject the null hypothesis and say that there is evidence that shows that this particular signal construction yields extreme returns.
