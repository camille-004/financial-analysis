# Load Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import datetime

# Visualization
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 6)

from pandas_datareader import data as pdr
yf.pdr_override()

start = datetime.datetime(2000, 1, 1)
end = datetime.datetime(2020, 1, 1)

tickers = [
    'AAPL', 'MSFT', 'AMGN', 'WMT', 'JNJ', 'JPM',
    'PG', 'UNH', 'HD', 'DIS', 'VZ', 'IBM',
    'KO', 'NKE', 'MRK', 'INTC', 'CSCO',
    'CVX', 'MCD', 'HON'
]

stocks = pdr.DataReader(tickers, start, end, interval='1mo')[['Adj Close', 'Volume']]

# Show only the first five rows for visual clarity
stocks.head()
```

# Calculate Monthly Returns

```python
monthly_returns = stocks['Adj Close'].pct_change()
monthly_returns.head()
```

```python
monthly_returns.shape
```

## 12-Month Moving Average

```python
ma = monthly_returns.rolling(12).mean()
ma.tail()
```

```python
ma.shape
```

# Equal-Weighted Portfolio
We have to create an extra DataFrame with weights by which to multiply the `monthly_returns` DataFrame so that we can avoid loss of information due to missing/`NaN` values.

```python
number_assets = len(stocks['Adj Close'].columns)
weight = 1 / number_assets
weight_df = monthly_returns.copy()
weight_df[pd.notnull(weight_df)] = 1
weight_df = weight_df.multiply(1 / weight_df.sum(axis=1), axis=0)
weight_df.head()
```

```python
monthly_returns['EWP'] = (weight_df * monthly_returns).sum(axis=1)
```

```python
monthly_returns['EWP']
```

```python
plt.title('Monthly Returns of Equal-Weighted Portfolio')
plt.xlabel('Year')
plt.ylabel('Equally-Weighted Monthly Returns Across Assets')
plt.plot(monthly_returns['EWP'], lw=1)
plt.plot(monthly_returns['EWP'].rolling(12).mean(), label='12-Month Moving Average')
plt.legend()
```

# EWP Excluding Monthly Bottom Deciles

```python
vol = stocks['Volume']
vol.head()
```

Fill in the missing values with the $n$-th row's mean.

```python
vol = vol.T.fillna(vol.mean(axis=1)).T
vol.shape
```

List including values, per row, of stocks that fall in the bottom decile.

```python
vol['10th_pct'] = vol.apply(lambda x: np.percentile(x, 10), axis=1)

def get_bottom_decile_stocks_per_mo(df):
    final = []
    for i in range(len(df)):
        row_df = df.iloc[i]
        decile_bool = (
            row_df <= row_df['10th_pct']).drop(labels=['10th_pct'])
        final.append(list(decile_bool[decile_bool == True].index))
    return final
     
deciles_per_mo = get_bottom_decile_stocks_per_mo(vol)
deciles_per_mo[:10]
```

Now, replace the monthly lower decile stocks with `NaN`.

```python
price_df = stocks['Adj Close']

for i in range(len(price_df)):
    row_df = price_df.iloc[i]
    row_df[deciles_per_mo[i]] = np.nan
    
price_df.head()
```

## Create and Plot EWP with New DataFrame without Monthly Bottom Deciles

```python
monthly_returns_90_pctl = price_df.pct_change()

number_assets = len(stocks['Adj Close'].columns)
weight = 1 / number_assets
weight_df_90 = monthly_returns_90_pctl.copy()
weight_df_90[pd.notnull(weight_df_90)] = 1
weight_df_90 = weight_df_90.multiply(1 / weight_df_90.sum(axis=1), axis=0)

monthly_returns_90_pctl['EWP'] = (weight_df_90 * monthly_returns_90_pctl).sum(axis=1)
ewp = monthly_returns_90_pctl['EWP']

plt.title('Monthly Returns of Equal-Weighted Portfolio')
plt.xlabel('Year')
plt.ylabel('Equally-Weighted Monthly Returns Across Assets')
plt.plot(ewp, lw=1)
plt.plot(ewp.rolling(12).mean(), label='12-Month Moving Average')
plt.legend()
```

## A Quick Comparison of Moving Average with and without Monthly Deciles Dropped
There will be a loss of accuracy in the second calculation with the row-by-row means imputed in the `vol` DataFrame.

```python
plt.plot(monthly_returns['EWP'].rolling(12).mean(), label='Deciles Not Dropped')
plt.plot(ewp.rolling(12).mean(), label='Deciles Dropped')
plt.legend()
```
